In this lab I look at how different factors effect crime rates in 1960 across 47 states in the United States. Using regression, I look at graphs and the R value to see the correlation. I chose this data set because I am very interested in criminology, and wanted to see what factors were correlated with crime rates in 1960. I had preconceived notions about what I would conclude from this lab, and was surprised by the results.

```python
#Wealth Versus Crime

import matplotlib.pyplot as plot
% matplotlib inline
import scipy.stats as stats
import numpy as np

file = open("Final.csv", "r", errors="ignore")
data =[]
for row in file:
    data.append(row)
del data[0]
for row in range(len(data)):
    data[row]=data[row].split(",")

wealth = []
crime = []

for row in range(len(data)):
    stateData = data[row]
    wealth.append(int(stateData[11]))
    crime.append(int(stateData[15]))
plot.scatter(wealth,crime)
plot.title("Wealth Versus Crime")
plot.xlabel("Wealth")
plot.ylabel("Crime")

print(stats.linregress(wealth, crime))

x = np.arange(0, 7500, 5);
y = 0.1768*x+-24.2826
plot.plot(x, y)
```

This graph, looking at wealth versus the crime rate, is a bit surprising. So the variable wealth, is essentially the median value of family income, and crime rate is the number of offenses committed per 100,000 people. The graph shows that the higher the median family income in a state, the higher the crime rate. Most people think that the poorer the people are, the more crime that takes place. I thought this is what this graph would show, and was surprised to see that it shows otherwise. This could be due to many factors. The R value is considerably low, being 0.4413, so the conclusions that can be made from this graph are not too reliable.
